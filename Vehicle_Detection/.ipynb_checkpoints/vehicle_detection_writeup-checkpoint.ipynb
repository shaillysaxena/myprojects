# Vehicle Detection Project 5

## Goals  
Goals of the project are as follows:  
1. Extract features from an image using HOG features, color transforms and binned color features.
2. Train a linear SVM on these features to classify cars from non-cars.
3. Implement a sliding-window technique and use this trained classifier to search for vehicles in the image.
4. Run this pipeline on video stream and create a heat map of recurring detections frame by frame to reject outliers and follow detected vehicles.
5. Estimate a bounding box for vehicles detected.

I am using vehicle and non-vehicle data provided by Udacity and most of the code is taken from lectures.

### 2. Explain how you settled on your final choice of HOG parameters.
Finding HOG parameters was based on experimentation with various parameters in skimage.hog(). I first started with RGB color space, moved to HSV and rested on YCrCb because it gave the best results after iterations of feature parameter tweaking, visualizing images and checking the effect of video.  
I also added color histogram and spatial binning to the features because adding these gave a very strong detection of cars and non-cars images.  

The final parameters obtained are described in the dictionary below:  

features_paramaters = {'color_space': 'YCrCb',
                       'orient': 9,
                       'pix_per_cell': 8,
                       'cell_per_block': 2,
                       'hog_channel': 'ALL',
                       'spatial_size': (32, 32),
                       'hist_bins': 32,
                       'spatial_feat': True,
                       'hist_feat': True,
                       'hog_feat': True}  
All these values came after experimentation.  

The final features are 8460 in length and this number, though high, works very well in doing the task of classification. I experimented with higher values of orient (30, 24, 16) and reached a good result at lower values with combination of other features.

## Sliding Window Search  
### 1. Describe how and where in your code you implemented a sliding window search. How did you decide what scales to search and how much to overlap windows?  

For sliding windows, code is available in slide_windows and search_windows functions in notebook 'vehicle_detection'. The general approach is to slide the windows of a fixed size over the area in the image that is highly likely to find a car. I only take pixels in y range 400-600 and disregard the first 200 pixels in the x range for this reason. An overlap of 0.7 for every window gives good results though a little time consuming.  
I experimented with window sizes of 64, 96 and 128 and the number of false positives increases as we increase the size of windows. On increasing the threshold for heatmap, this problem was overcome. After experiments, it was found that window size of 96 is a good tradeoff between speed and efficiency where window size of 64 is too slow and 128 gives a lot of false positives.


### 2. Show some examples of test images to demonstrate how your pipeline is working. What did you do to optimize the performance of your classifier?
I had first started with only HOG features but the number of false positives was high. I over-sampled the non-cars data so that the classifier learns non-cars data properly because this is the kind of data the classifier is going to see more often.  
I then added spatial binning and color histogram features and the resutls improved a lot. In order to increase the processing speed, I increased the values of window sizes and the number of false positives increased again. So I had to play around with heatmap thresholding and accumulation of heatmaps in order to achieve a desirable result.
In the pipeline, I followed these steps:  
1. Extract features from cars and non-cars data
2. Slide windows over the region of interest of the image and find hot windows where the cars are found.  
3. Keep a list of all the hot windows for 60 frames so that we can get a smooth bounding box for the detected cars. And remove the last few frames when this list is full.
4. Convert the image to its heaptmap and threshold it so that the false positives are reduced. This is based on the fact that the detected cars will be available in more frames than false positives.

### 2. Describe how and where in your code you implemented some kind of filter for false positives and some method for combining overlapping bounding boxes.  
I recorded the positions of positive detections in each video frame of the video. From the positive detections, I created a heatmap and then thresholded the map to identify the vehicle positions. I then used scipy.ndimage.measuremenst.label() to identify the individual blobs in the heatmap. I then assumed that each blob corresponded to a vehicle. I constructed bounding boxes to cover the area of each blob detected.  
On increasing the window size, I found out that the number of false positives increased and hence, I experimented with the heatmap thresholds. Finally, found out the point where threshold and false positives could be balanced.  
This code is available in 'vehicle_detection' notebook under the function 'pipeline'.

## Discussion

### 1. Briefly discuss any problems / issues you faced in your implementation of this project. Where will your pipeline likely fail? What could you do to make it more robust?  
In this project, the biggest challenge was to get rid of the false positives in the video. Even though the accuracy of classifier is high, it is not very clear if its performance in the video will be good enough. You can only tell at the end of the pipeline if things worked. I had started with only HOG features but later added other features too which helped a lot. There are a few things I would have liked to add to my project:  
1. Faster implementation of extracting HOG features by using subsampling.
2. Feature selection using decision trees for a smaller feature set.
3. Windows of different sizes to detect cars, currently I am using a fixed size for window search.
4. The area of interest is hardcoded and I think is not too reliable in varying situations like slopes.  

Problem areas where this pipeline could fail are in conditions of slopes or wide variety of false positives that the classifier could predict. It will help to add more non-vehicle images of things present on the road for the classifier to train on.
To make this more robust, I could try learning more about Single Shot Multibox Detector or YOLO (You Only Look Once) for a faster and more robust implementation of vehicle detection.
