# Behavioral Cloning Project - Report

## Goals
The goal of this project is to drive the car in the simulator successfully without any of the tires touching the non-drivable portion of the road. We will achieve this goal by making use of Convolutional Neural Networks.

## Files Submitted
Please find files model.py, model.h5, model.json, behavioral_cloning.ipynb, video.mp4 and drive.py which have been created/transformed according to my implementation of the problem. behavioral_cloning.ipynb is the file where essentialy all the implementation is done. This is also available in model.py. I used Udacity's data for this project.  

In order to run the project, type 'python drive.py model.json' from the terminal.  



## Input Data

Training data has been taken from Udacity and I have used both left and right images with steering correction of 0.3. I also flip the image and its steering angle to create more meaningful data.  

For the final architecture I chose, I perform the follwing pre-processing steps to each image:  
1. Convert the image from RGB to HSV
2. Extract the S channel from the image which is the most differtiable channel when it comes to areas in the road which are most problamatic to drive on.
3. Crop the image in a way that learning is independent of trees and sky at the top of the image.
4. Resize the image to 16 rows and 32 columns which makes dealing with 24k+ images quickly and efficiently.

## Network Architecture

In order to reduce overfitting, I use dropout of 0.25 which works well. Also, I use validation set of 10% from the image data after shuffling.  

ReLU layers introduce non-linearity and I used Adam optimizer so manual tuning of learning rate was not necessary.  

Mean squared error is chosen to be the loss function to be minimized but it is not a good representative of what will happen on the track. So, testing is done after saving every new model on the track so as to see how the model is performing.  

A Lambda layer to normalize the input image is also used.

### Training Process

I trained the model for 10 epochs and a batch size of 128 but these values came after experimentation. In case of augmentation I was training the model on 5 epochs with a batch size of 20000 which was fruitless and very time consuming even on AWS GPU instance. I could train my final model on my own machine due to the small size of the input images.
